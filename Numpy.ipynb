```python
a1 = np.arange(0, 10)
a2 = np.arange(0, 10).reshape(5, 2)
a3 = np.arange(0,10).reshape(1,2,5)
```

```python
#ndim - what is the dimension of the array o the data
print(a1.ndim)
print(a2.ndim)
print(a3.ndim)
```

```python
#shape - tells what is the shape of the data
print(a1.shape)
print(a2.shape)
print(a3.shape)
a3
```

```python
#size - total values in the data
print(a1.size)
print(a2.size)
print(a3.size)
```

```python
#itemsize - tells what is the size of datatype used (float - 8 bytes)
print(a1.itemsize)
#dtype - tells waht data type is the data
```

```python
#astype - change the data type
a1.astype('float')
print(a1)
```

```python
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)
print(a1)
print(a2)
```

```python
##Scalar Operation ##

#arthimetic ( +, -, /, **, %)
a1 % 2
```

```python
## Relational Operation ## ( >, <, !=, ==, >=, <= )
a2 != 5
```

```python
## Vector Operations ## ( Same as scalar operations but between vectors)

a1 + a2
```

```python
## Array Functions ##

a1 = np.random.random((3, 5))
a2 = np.round(a1 * 100)
print(a1, '\n', a2)
```

```python
np.sum(a1)
np.prod(a2)
np.min(a1)
np.max(a2)
np.min(a1, axis= 1) # axis = 1 for rows and 0 for col
np.mean(a1)
np.median(a2)
np.std(a1)
np.var(a2)
a1 = a1.reshape(5, 3)
np.dot(a1, a2)
np.exp(a1)
np.log(a2)
```
